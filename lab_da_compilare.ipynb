### Apprendimento supervisionato: classificazione

Si tratta di una tecnica per riuscire a scrivere una funzione *y=f(x)* in grado di fornire un'etichetta (es. dove *y* è A o B, o 0 o 1, ...) per una *x* in input.

Per trovare *f*, bisogna avere un dataset di dati storici di coppie (x, y) e un algoritmo in grado di apprendere *f(x)* a partire dai dati.

Consideriamo il dataset *fever2.csv* e proviamo a trovare la migliore linea che divida le persone con la febbre e quelle senza.

Definiamo una funzione chiamata *predici_febbre* in grado di calcolare *d = aT+bH+c*. *H*, *T*, *a*, *b* e *c* sono tutti input della funzione. Data una certa coppia di valori *H* e *T*, se *d < 0*, allora diremo che la persona ha la febbre, altrimenti diremo senza febbre. La funzione deve calcolare  *d=aT+bH+c* e il suo output deve essere 1 se d<0 o 0 altrimenti.

Creiamo la funzione e usiamo *a=-1*, *b=-1* e *c=45*.

Scriviamo:
```python
def predici_febbre(H, T, a, b, c):
    d = a*T + b*H + c
    return (d < 0) + 0.0

a = -1
b = -1
c = 45

Fp = predici_febbre(H, T, a, b, c)
print(Fp)
print(F)
```


Calcolare l'accuratezza della classificazione usando la funzione calcola_accuratezza(F, Fp)

Scrivere:
```python
acc = calcola_accuratezza(F, Fp)
print(acc)
```

Dire *d < 0* -> febbre, significa dire "tutti i punti (H, T)* tale che *aT+bH+c>0*. Risolvendo la disequazione si ottiene una retta del tipo: *T = -(b/a)H - (c/a) = m H + q* con *m=-b/a* e *q=-c/a*. Questa retta è chiamata **decision boundary**.

Eseguire questi passaggi:
- Definire la funzione *retta(H, a, b, c)* in grado di calcolare *T = m H + q*
- Creare un range di *H* da testare e provare a predire il valore di *T* con la funzione appena creata. Chiamare il range *Hrange* e la predizione *Trange*.
- Ridisegnare lo scatter plot tra *H* e *T* del dataset
- Plot della retta con *Hrange* e *Trange*

Scrivere:
```python
def retta(H, a, b, c):
    m = -(b/a)
    q = -c/a
    T = m*H + q
    return T

Hrange = np.arange(6, 9, 0.1)
Trange = retta(Hrange, a, b, c)
```
